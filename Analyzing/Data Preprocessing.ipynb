```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('DataSets/movies.csv')
credits=pd.read_csv('DataSets/credits.csv')
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
movies.head()
```

```python
credits.head()
```

```python
df=movies.merge(credits,on='title')
```

```python
# df.head()
```

```python
movies_df=df[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies_df.head()
```

```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_df['genres']=movies_df['genres'].apply(convert)
movies_df['keywords']=movies_df['keywords'].apply(convert)
```

```python
movies_df.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies_df['cast']=movies_df['cast'].apply(convert3)
```

```python
movies_df.head()
```

```python
def fetchDirector(obj):
    L=[]
    for i in ast.literal_eval(obj):
        
        try:
            if i['job'].lower()=='director':
                L.append(i['name'])
                break
        except:
            L.append(None)
    return L
```

```python
movies_df['crew']=movies_df['crew'].apply(fetchDirector)
```

```python
movies_df.head()
```

```python
movies_df.to_csv('Analyzed_DataSets/movies_data.csv')
```
